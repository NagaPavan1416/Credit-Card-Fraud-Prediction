```python
df_num=df.iloc[0:,0:30]
df_num
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df_num=ss.fit_transform(df_num)
df_num=pd.DataFrame(df_num)
df_num.columns = list(df_num)
df_num.head()
```

```python
Y=df['Class']
X=df_num
```

```python
#data particition

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.20)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(x_train, y_train)
```

```python
#predicting
y_pred=LR.predict(x_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score
cm=confusion_matrix(y_train,y_pred)
print(cm)
print('accuracy_score:',accuracy_score(y_train,y_pred))
print('recall_score:',recall_score(y_train,y_pred).round(2))
print('precision_score:',precision_score(y_train,y_pred).round(2))
print('F1_score:',f1_score(y_train,y_pred).round(2))
```
